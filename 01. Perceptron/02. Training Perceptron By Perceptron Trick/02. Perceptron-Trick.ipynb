# Perceptron Trick

Perceptron trick is a shortcut method to train a perceptron that is very simple and easy to understand. It very helpful in making us understand the intution behind the training of a perceptron.
Here, in this part we will get to know the behaviour of a Perceptron during its training.

# Working

We, already know that our objective behind trainig is to calculate the the values of our Weights and Bias to find the equation of the final line. <br>
Assume that the line is &nbsp; <b> Ax<sub>1</sub> + Bx<sub>2</sub> + C = 0 </b>

1. Start with random values of <b>A</b>, <b>B</b> and <b>C</b>
    1. Now we will select a random point from our data and calculate the value of <b>Z</b> for it. [ Passing points in &Sigma; function ]
    2. Now, according to the calculated value of <b>Z</b> the point will either fall in the `+ive ( Z >= 0 ) or -ive ( Z < 0 )` region.
    3. Now we will check weather our prediction matches with the category of the input or not.
    4. * IF matches : we will continue this loop.
       * ELSE : we will change the values of <b>A</b>, <b>B</b> and <b>C</b> accordingly and will continue with new values. <b> [ Transformation ]</b>
    5. And this loop will continue until the Perceptron Converges<sup>*</sup> or the Number Of Epoches is reached.

<b> Perceptron Convergence </b> : `when num( Misclassified Points ) = 0`

# Transformation
There are two cases in which we need to transform the line. The transformations for there two cases are dscribed below :

1. `Case 1: Belongs To -ive, Predicted +ive` <br>
   * Append 1 to the cordinates of the selected point. <br>
     For ex: `(3, 5) => (3, 5, 1)`
   * Subtract the cordinates of the point multiplied by <b>&eta;</b> from the Values of A, B and C. <br> 
     For ex: `(A, B, C) => (A-3η, B-5η, C-1η)`
<br>
<br>
2. `Case 2: Belongs To +ive, Predicted -ive` <br>
   * Append 1 to the cordinates of the selected point. <br>
     For ex: `(3, 5) => (3, 5, 1)`
   * Add the cordinates of the point multiplied by <b>&eta;</b> to the Values of A, B and C. <br>
     For ex: `(A, B, C) => (A+3η, B+5η, C+1η)`

<b>Learning Rate Or η (Eta)</b> : `a small number generally 0.1 or 0.01 to keep transformation changes small`

# Simplifying General Equation

Now we have : `Ax + By + C = 0` <br>
Rewrite it as &nbsp; : `W₀ + W₁X₁ + W₂X₂ = 0` <br>
This becomes : `W₀X₀ + W₁X₁ + W₂X₂ = 0` <br>

Finally we get : &nbsp; &nbsp; $\sum_{i=0}^{2}{WᵢXᵢ} = 0$ <br>
where, W₀ is the bias, and X₀ is a column we append in dataset with all values 1 to get Simplify the equation, rest values are the remaining weight, and their corresponding input pairs.

Therefore the final general formula becomes : &nbsp; &nbsp; $\sum_{i=0}^{n}{WᵢXᵢ} = 0$ <br>

<b>Which is also the dot product of i<sup>th</sup> (1 X n) data row with the (n X 1) matrix having all values of W.</b>

# Implementing Perceptron Trick

```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, 
                         random_state=41, hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.scatter( X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def step(z):
    return 1 if z >= 0 else 0

def Perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(5000):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j]-y_hat) * X[j]

    return weights[0], weights[1:]
```

```python
intercept_, coef_ = Perceptron(X, y)
print(intercept_)  # C
print(coef_)       # [A, B]  for  Ax + By + C = 0
```

```python
# Convetring it from  Ax + By + C = 0   To   Y = mX + b
m = - (coef_[0] / coef_[1])    # m = -A/B
b = - (intercept_ / coef_[1])  # b = -C/B 
```

```python
# Plotting line on the Scatterchart
x_input = np.linspace( -3, 3, 100)
y_input = m * x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter( X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```

# Animation For Visualization Perceptron Trick

```python
def animate_perceptron(X,y):
    m, b = [], []
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(100):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j]-y_hat) * X[j]

        m.append(-(weights[1]/weights[2])) # m = -A/B
        b.append(-(weights[0]/weights[2])) # b = -C/B
    
    return m,b
```

```python
m, b = animate_perceptron(X, y)
```

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
import numpy as np

fig, ax = plt.subplots(figsize=(9,5))

x_i = np.arange(-3, 3, 0.1)
y_i = x_i * m[0] + b[0]
ax.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
line, = ax.plot(x_i, x_i * m[0] + b[0], 'r-',linewidth=2)
plt.ylim(-3, 3)

def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i * m[i] + b[i])
    ax.set_xlabel(label)


show_animation = FuncAnimation(fig, update, repeat=True, frames=100, interval=200)
show_animation.save("02. perceptron.gif", writer="pillow")
```
